![ML pipeline](chap02_figs/ml_project.png)

<div class="alert alert-block alert-info">
    <center><b>Pipeline</b></center><br>

一系列的数据处理组件 (data processing components) 称之为 <b>pipeline</b>。 pipeline 在 ML 系统很常见，因为 ML 需要大量的<i>数据操作 (data manipulation)</i> 和<i>数据转换 (data transformation)</i> 才可以应用。

<br>

<b>各组件通常是异步 (asynchronously) 运行的。</b> 每个组件都 pull in 数据，处理数据，最后将结果传到另一个 data store 中。然后，下一个组件获取前面组件的结果，并给出自己的结果，以此类推。<b>每个组件都很独立 (self-contained)：组件与组件之间的连接只通过 data store。</b> 这使得系统简单易懂 (simple to grasp)，不同团队专注不同组件即可。而且如果某个组件发生故障 (break down)，下游组件可以使用前一个组件的输出继续运行一段时间，所以整体结构的鲁棒性 (roubust) 较强。

<br>

当然，如果没有适当的监控，坏的组件未被发现，那么过期的 (stale) 数据会使得整个系统性能下降。
</div>

## 2.2 选择性能指标

**下一步是选择一个性能指标 (performance measure)。**

**<font color='crimson'>回归问题的典型性能指标是 RMSE (Root Mean Square Error)，给出了算法通常会在预测中产生多大的误差，误差越大，权重越高。</font>**

$$RMSE(X, h)=\sqrt{\frac{1}{m}\sum^m_{i=1}(h(x^i)-y^i)^2}$$

虽然，RMSE 通常是回归任务的首选性能指标。但在某些情况下，可能需要其他指标。**<font color='crimson'>在异常值 (outlier) 比较多的时候，考虑使用 MAE </font>** (Mean Absolute Error, also called Average Absolute deviation)。

$$MAE(X, h) = \frac{1}{m}\sum^m_{i=1}|h(x^i)-y^i|$$

RMSE 和 MAE 都是测量两个向量 (vector) 之间距离的方法，其他可用的距离度量方式或者范数 (norm)：

- RMSE 与欧几里得范数 (Euclidean norm) 相对应，也称为 **<font color='blue'>l2范数</font>**，记为 $||.||_2$ 或者 $||.||$。


- MAE 对应于 **<font color='blue'>l1 范数</font>**，记为 $||.||_1$。有时又称为 Manhattan norm，因为在计算城市中 2 个点距离时，只能沿着正交的街区行动。


- 一般而言，包含 n 个元素的向量 v 的 $l_k$范数定义为 $||v||_k=(|v_0|^k+|v_1|^k+\cdots+|v_n|^k)^{1/k}$。


- $l_0$ 范数给出了向量中非 0 元素的数量，$l_\infty$范数是向量中最大元素绝对值，$||v||_\infty=max(|v_0|, |v_1|, ..., |v_n|)$。


- **<font color='crimson'>范数的 index 越大，范数对较大值越敏感 (focus on large values)，而忽略较小值 (neglect small values) 。这也是 RMSE 比 MAE 对异常值更敏感的原因。但是，当异常值呈指数形式稀少的时候 (如钟形曲线 (bell-shaped curve))，RSME 表现更好，通常是首选。</font>**


- **<font color='blue'>Huber损失</font>**结合了 RMSE 和 MAE。


$$
L(y, f)=\begin{cases}
\frac{1}{2}(y-f)^2, |y-f|\leq \delta \\\\
\delta|y-f|-\frac{1}{2}\delta^2, |y-f|>\delta
\end{cases}
$$
> $\delta$ 通常设为 1。

```python
import os
import tarfile
import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join('datasets', 'housing')
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

## 3.4 留出测试集

<div class="alert alert-block alert-success">
    <b>总的来说，在切分数据集的时候（一定要留出测试集），使用 Scikit-Learn 的 <code>train_test_split()</code>，并且考虑在比较重要的（一个或多个）特征上做分层抽样。如果该特征是连续型的数值，需要将其转换成类别特征 (<code>pd.cut()</code> 做分箱），stratum 的数量不宜过多，且每个 stratum 中要有足够多的样本。</b>
</div>

**<font color='crimson'>留出测试集非常重要，为什么呢？</font>**

虽然在决定使用什么算法之前需要进一步探索数据集（到目前为止只简单看了一下），但是人的大脑是一个非常神奇的模式识别系统 (pattern detection system)，这意味着非常容易 overfitting：一旦看了测试集，就会注意到测试集中的有趣/特殊模式，然后就会选择某个特殊的 ML 算法。显然，这个算法在测试集上的表现非常好，但是在实际中就不行了。这就是 **<font color='red'>data snooping (窥探、探听) bias</font>**。

留出测试集理论上非常简单 (theoretically simple)，随机选择一些（通常 20%）样本放在一边就可以了。

````python
# For illustration only
def split_train_test(data, test_size=0.2):
    """Pick some samples randomly."""
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_size)

    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
````

````python
train_set, test_set = split_train_test(housing, 0.2)
train_set.shape, test_set.shape
````

虽然可行，但没法用。因为每次运行的结果不一样。这样下去的话，就会看到整个数据集了，而这正是需要避免的（留出测试集的原因）。有 2 种方法可以解决这个问题：

- 运行一次之后，立马保存，后面仅仅是加载测试集了。


- 在 `np.random.permutation` 之前运行 `np.random.seed(42)`。设置好随机数生成器种子 (random number generator's seed) 之后，就会生成相同的数据了。

但是，下一次获取更新的数据后，就会出现问题。**<font color='crimson'>为了在更新数据集后也有一个稳定训练集/测试集分割，常见的方法是每个样本都使用一个标识符 (identifier) 来决定是否分到测试集中（假定每个样本都有一个唯一且不变的标识符）。</font>** 如，计算出每个样本标识符的哈希值 (hash)，如果该哈希值 <= 最大哈希值的20%，那么将该样本分到测试集中。这样的话，多次运行的话测试集中的样本也是不变的，而且即使更新数据集，新增的数据中也有 20% 被分到测试集中，之前的训练集不会被分到测试集中。

````python
from zlib import crc32


# 支持 python2 和 python3
def test_set_check(identifier, test_size):
    return crc32(np.int64(identifier)) & 0xffffffff < test_size * 2**32


def split_train_test_by_id(data, test_size, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_size))
    return data.loc[~in_test_set], data.loc[in_test_set]
````

````python
import hashlib


# 支持任意哈希函数，但是比较慢，而且不支持 python2
def test_set_check(identifier, test_size, hash=hashlib.md5):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_size
````

````python
# 支持任意哈希函数，与 Python 2 和 Python 3 兼容
def test_set_check(identifier, test_size, hash=hashlib.md5):
    return bytearray(hash(np.int64(identifier)).digest())[-1] < 256 * test_size
````

但是，这个数据集没有标识符列。最简单的方法是用行的 index 作为标识符：

````python
housing_with_id = housing.reset_index()  # 添加一个标识符列
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'index')
train_set.shape, test_set.shape
````

**如果使用行的 index 作为样本标识符的话，需要确在原始数据集后面新增数据，而且不能删除原始数据任何一行。** 如果做不到的话，就尝试使用最为稳定的特征作来构建一个标识符。如，这里使用经纬度作为标识符。

````python
housing_with_id['id'] = housing['longitude'] * 1000 + housing['latitude']
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'id')
train_set.shape, test_set.shape
````

Scikit-Learn 中有很多方法来切分数据集。最简单的是使用 **`train_test_split()`**，与前面的 `split_train_test()` 类似，但是有一些额外特征：

- 参数 `random_state` 用来设置随机生成器种子 (random generator seed)


- 可以将行数相同的多个数据集（如 X 和 y）一起传入，然后根据相同的索引切分数据

````python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
train_set.shape, test_set.shape
````

````python
test_set.head(2)
````

如果数据集非常大，纯随机抽样 (purely random sampling) 没有任何问题，否则的话就会导致**<font color='red'>抽样偏差 (sampling bias)</font>** ，此时应该使用**<font color='red'>分层抽样 (stratified sampling)</font>**。

这里，`median_income` 是一个比较重要的特征。因此，希望测试集具有代表性——测试集中该特征各个 category 的比例与整体数据集中各个 category 的比例相同。但是，`median_income` 是连续型特征 (continuous numerical attribute)，因此，需要创建一个对应的类别特征 (category attribute)。

````python
plt.hist(housing['median_income'], bins=50)
# housing['median_income'].hist()
plt.xticks(np.arange(0, 16))
plt.show()
````

可以看出，`median_income` 聚集在 1.5～6 之间，也有一些值远远超过了 6。**<font color='crimson'>分层抽样的时候需要确保每一层  (stratum) 都有足够多的样本，否则的话，对该 stratum 的重要性估计会有偏差。不应该有太多的 strata，每个 stratum 中要有足够多的样本。</font>**

````python
housing['income_cat'] = pd.cut(housing['median_income'],
                               bins=[0, 1.5, 3., 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])
housing['income_cat'].hist()
plt.show()
````

````python
# 基于 income_cat 作分层抽样
from sklearn.model_selection import train_test_split

strat_train_set, strat_test_set = train_test_split(
    housing,
    test_size=0.2,
    stratify=housing['income_cat'],  # stratified sampling
    random_state=42)
````

```python
# 原文代码
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

````python
# 测试集中 income_cat 各个值的比例
strat_test_set['income_cat'].value_counts(normalize=True).sort_index()
````

````python
# 完整数据集中 income_cat 各个值的比例
housing['income_cat'].value_counts(normalize=True).sort_index()
````

````python
# Random split
random_train_set, random_test_set = train_test_split(
    housing, test_size=0.2, random_state=42)
````

````python
# 比较一下
compare_props = pd.DataFrame({
    'Overall': housing['income_cat'].value_counts(normalize=True).sort_index(),
    'Stratified': strat_test_set['income_cat'].value_counts(normalize=True).sort_index(),
    'Random': random_test_set['income_cat'].value_counts(normalize=True).sort_index(),
})
compare_props['Random %error'] = (
    100 * compare_props['Random'] / compare_props['Overall']
    - 100)
compare_props['Stratified %error'] = (
    100 * compare_props['Stratified'] / compare_props['Overall']
    - 100)
compare_props
````

可以看出，分层抽样的测试集中比例分布与原始完成数据集基本是一样的，而随机抽样是有偏的 (skewed)。

````python
# 删除用于分层抽样的辅助特征——income_cat
for set_ in (strat_train_set, strat_test_set):
    set_.drop(columns='income_cat', inplace=True)
````

虽然花了很多时间，但是非常值得，因为这个非常重要，但是经常被忽略。另外，这里的一些想法在后面讨论交叉验证 (cross validation) 也有用。

## 4.2 相关性（Correlation）

数据不是很大，所以可以使用 **`corr()`** 计算特征之间的**<font color='red'>标准相关系数 (standard correlation coefficient, also called Pearson's r)</font>**。

````python
# 方法1：特征之间的相关性
corr_matrix = housing.corr()
corr_matrix
````

````python
# 每个特征与 median_housing_value 之间的相关性
corr_matrix['median_house_value']
````

**<font color='crimson'>相关系数的范围是从 -1 到 1</font>**

- 越近 1，表示越强的正相关性 (positive correlation)（如，你增加，我增加）


- 接近 -1 时，表示有较强的负相关性 (negative correlation)（如，你增加，我减少）


- 接近 0 时，两者之间不存在线性相关 (linear correlation)（如，你是你，我是我）

<br>

![线性相关](chap02_figs/standard_correlation_coefficient.png)

<div class="alert alert-block alert-danger">
- <b>相关系数只能衡量线性相关性 (linear correlation)</b>，如 y 随着 x 增加而增加/减少。但是，<b>这可能忽略非线性关系 (nonlinear relationships)</b>（如，随着 x 接近 0，y 增加）。<b>上图中的最后一行，相关系数为 0，但是存在非线性关系。</b>
    <br><br>
- <b>上图第二行，相关系数为 1 或 -1，这和斜率 (slope) 无关。</b>
</div>
